```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
```

```python
def scrape_bbref(url, title):
    page = requests.get(url, timeout=5)
    html_content = page.text
    soup = BeautifulSoup(html_content, 'lxml')
    table_row = soup.find_all('tr', class_='full_table')
    # isolate table rows in the class 'full_table'table_row = soup.find_all('tr', class_='full_table')
    list_of_rows = []
    for each_row in table_row:
        td = each_row.find_all('td')
        row = [i.text for i in td]
        list_of_rows.append(row)
    table_headers = soup.find_all('th', class_='poptip')
    headers = [i.text for i in table_headers]
    df = pd.DataFrame(list_of_rows)
    df.columns = headers[1:]
    df['season_start'] = int(url[49:53])-1
    df['season_end'] = url[49:53]
    df.to_csv(r'../data_historical_basic_stats/' + title + '_' + url[49:53] + '.csv', index=False)
#     df.to_csv(title + f'{datetime.now()}' + '.csv', index=False)
```

```python
def scrape_mvp(url):
    page = requests.get(url, timeout=5)
    html_content = page.text
    soup = BeautifulSoup(html_content, 'lxml')
    soup_table = soup.find('table', class_='sortable stats_table')
    soup_tr = soup_table.find_all('tr', class_=None)
    # isolate table rows in the class 'full_table'table_row = soup.find_all('tr', class_='full_table')
    list_of_rows = []
    for each_row in soup_tr:
        td = each_row.find_all('td')
        row = [i.text for i in td]
        list_of_rows.append(row)
    soup_headers = soup_table.find_all('th', class_='poptip')
    headers = [i.text for i in soup_headers]
    df = pd.DataFrame(list_of_rows)
    df.columns = headers[1:]
    df.drop(index=0, inplace=True)
    df['mvp_year'] = url[51:55]
    df.to_csv(r'../mvp_data/' + 'mvp' + '_' + url[51:55] + '.csv', index=False)
```

```python
def scrape_winners(url):
    page = requests.get(url, timeout=5)
    html_content = page.text
    soup = BeautifulSoup(html_content, 'lxml')
    soup_table = soup.find_all('table', class_='suppress_all sortable stats_table')
    columns = ['wins', 'losses', 'win-loss', 'games_behind', 'ps/g', 'pa/g', 'srs']
    appended_df = []
    for i in range(2):
        soup_tr = soup_table[i].find_all('tr', class_='full_table')
        list_of_rows = []
        for each_row in soup_tr:
            td = each_row.find_all('td')
            row = [i.text for i in td]
            list_of_rows.append(row)
        teams = [i.text for i in soup_tr]
        df = pd.DataFrame(list_of_rows, columns=columns)
        df['team'] = teams
        appended_df.append(df)
    df = pd.concat(appended_df).reset_index(drop=True)
    df['team'] = df['team'].apply(lambda x: x.split('\xa0')[0].split('*')[0])
    df['season_start'] = int(url[49:53])-1
    df['season_end'] = url[49:53]
    df.to_csv(r'../data_historical_win_record/' + 'team_record' + '_' + url[49:53] + '.csv', index=False)
```
