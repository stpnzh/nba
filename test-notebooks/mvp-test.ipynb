```python
testdf.drop(columns=['Rk', 'Player', 'Tm', 'GS', 'Pos', 'Season'], inplace=True)
```

```python
testdf.fillna(0, inplace=True)
```

```python
df2 = testdf[(testdf['G'] > 70) & (testdf['MP'] >= 30) & (testdf['PTS'] >= 15)]
```

```python
df2.drop(columns=['Age', 'G', 'eFG%', 'FG', 'FGA', 'FG%', '3P', '3PA'
                  , '3P%', '2P', '2PA', '2P%', 'FT', 'FTA', 'ORB', 'DRB'
                  , 'STL', 'BLK', 'TOV', 'PF'], inplace=True)
```

```python
y = df2['MVP']
X = df2.loc[:,df2.columns!='MVP']
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB # Gaussian Naive Bays
from sklearn.linear_model import Perceptron
from sklearn.tree import DecisionTreeClassifier
```

```python
logreg = LogisticRegression()                                # instantiate
logreg.fit(X_train, y_train)                                 # fit
y_pred = logreg.predict(X_test)                               # predict
acc_logreg = sum(y_pred == y_test)/len(y_test)*100                # evaluate

print('Logistic Regression labeling accuracy:', str(round(acc_logreg,2)),'%')
```

```python
# we could also use scikit learn's method score
# that predicts and then compares to validation set labels
acc_log_2 = logreg.score(X_test, y_test)                       # evaluate

print('Logistic Regression using built-in method:', str(round(acc_log_2*100,2)),'%')
```

```python
# Random Forest
random_forest = RandomForestClassifier(n_estimators=500)   # instantiate
random_forest.fit(X_train, y_train)                         # fit
acc_rf = random_forest.score(X_test, y_test)                  # predict + evaluate

print('randome forest:', str(round(acc_rf*100,2)),'%')
```
